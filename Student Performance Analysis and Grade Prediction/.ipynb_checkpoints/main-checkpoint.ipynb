```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("data/student_performance.csv")
#Data Cleansing
# Data Understanding & Inspection
print(f"Df Shape : {df.shape}")
print("Column Names : ")
print(df.columns)
print("Column Information")
df.info()
Selected_df=df.copy()
#Noise Reduction
Selected_df=Selected_df.drop(columns=["student_id","gender","school_type","parental_education","internet_access","weekend_alcohol_1_5","family_support","free_time_1_5","passed","extracurricular",	"health_1_5"	,"weekday_alcohol_1_5"]).copy()
independent_variables=Selected_df.iloc[:,0:7].copy()
independent_variables.head()
dependent_variable=Selected_df["final_grade_0_100"].reset_index()
Selected_df.describe()
'''
We are going to perform analysis to check Student Performance Factor 

'''
before_rows=Selected_df.shape[0]
Selected_df.dropna(inplace=True)
after_rows=Selected_df.shape[0]
print(f" No of Rows removed :\n { before_rows-after_rows }")
print(" Check if Data types need conversion ")
print(Selected_df.dtypes)


```

```python
col=Selected_df.select_dtypes(include="number").columns
for key in col:
    key_mean=Selected_df[key].mean()
    key_median=Selected_df[key].median()
    key_mode=Selected_df[key].mode()[0]
    key_variance=Selected_df[key].var()
    key_std=Selected_df[key].std()
    print(f"for {key }\n Mean : {key_mean:.2f}\n Median :{key_median:.2f} \n Mode {key_mode:.2f} \n Var {key_variance:.2f}\n std {key_std:.2f}")
```

```python
numeric_cols=Selected_df.select_dtypes(include="number").columns
for cols in numeric_cols:
    Q1=Selected_df[cols].quantile(0.25)
    Q3=Selected_df[cols].quantile(0.75)
    IQR=Q3-Q1
    lower=Q1-1.5*IQR
    upper=Q3+1.5*IQR
    outlier=Selected_df[(Selected_df[cols]<lower)|(Selected_df[cols]>upper)]
    print(f"So the Outliers we have in {cols}  is {outlier.shape[0]}")
```

```python
def saveandshowhist(colname,title,file_path):
    sns.histplot(x=colname,data=Selected_df)
    plt.title(title)
    plt.savefig(file_path)
    plt.show()


col_names=Selected_df.loc[:,["study_time_hours_per_week","math_score_0_100","reading_score_0_100","writing_score_0_100","final_grade_0_100"]].columns
title_all=["Distribution of Students by study_time_hours_per_week  ","Distribution of Students by  math score ","Distribution of Students by reading score","Distribution of Students by writing score","Distribution of Students by final score"]
file_path=["Graphs/Histograms/study_time_hours_per_week.png","Graphs/Histograms/math_score_0_100.png","Graphs/Histograms/reading_score_0_100.png","Graphs/Histograms/writing_score_0_100.png","Graphs/Histograms/final_grade_0_100.png"]

for x,title,filepath in zip(col_names,title_all,file_path):
    saveandshowhist(x,title,filepath)

```

```python
def saveandshowplot(colname,title,label1,label2,file_path):
    sns.countplot(x=colname,data=Selected_df)
    plt.title(title)
    plt.ylabel(label1)
    plt.xlabel(label2)
    plt.savefig(file_path)
    plt.show()


col_names=Selected_df.loc[:,["age","failures","absences"]].columns
title_all=["Distribution of Students age groups","Distribution of Students by failures","Distribution of Students by their absences"]
ylabel_all=["Number of students","Number of students","Number of students"]
xlabel_all=["Age group","failures","absences"]
file_path=["Graphs/Distribution/Age_groups.png","Graphs/Distribution/failures_groups.png","Graphs/Distribution/absences_groups.png"]

for x,title,labely,labelx,filepath in zip(col_names,title_all,ylabel_all,xlabel_all,file_path):
    saveandshowplot(x,title,labely,labelx,filepath)

```

```python
print("Checking Correlation ")
print("study_time_hours_per_week,final_grade_0_100")
print(Selected_df[["study_time_hours_per_week","final_grade_0_100"]].corr())
print("absences,final_grade_0_100")
print(Selected_df[["absences","final_grade_0_100"]].corr())
print("math_score_0_100,final_grade_0_100")
print(Selected_df[["math_score_0_100","final_grade_0_100"]].corr())
print("reading_score_0_100,final_grade_0_100")
print(Selected_df[["reading_score_0_100","final_grade_0_100"]].corr())
print("writing_score_0_100,final_grade_0_100")
print(Selected_df[["writing_score_0_100","final_grade_0_100"]].corr())
```



The correlation analysis indicates that academic subject scores (math, reading, and writing) have a strong positive relationship with final grades, whereas study time shows a weak positive association and absences exhibit a weak negative relationship. This suggests that subject-wise performance is a stronger indicator of final grades than behavioral factors such as study time or attendance.
